21. Bagaimana cara meringkas jumlah unik customer, jumlah unik invoice number, jumlah quantity, jumlah revenue, jumlah unik deskripsi, mengurutkan price dari terkecil ke terbesar, dan mengurutkan invoice price dari terbesar ke terkecil? Gunakan `def ... condition`

```python
def summarize_data(data):
    # Jumlah unik customer
    unique_customers = data['CustomerID'].nunique()

    # Jumlah unik invoice number
    unique_invoices = data['InvoiceNo'].nunique()

    # Jumlah quantity total
    total_quantity = data['Quantity'].sum()

    # Jumlah revenue
    total_revenue = (data['Quantity'] * data['UnitPrice']).sum()

    # Jumlah unik deskripsi
    unique_descriptions = data['Description'].nunique()

    # Mengurutkan price dari terkecil ke terbesar
    sorted_prices = data.sort_values(by='UnitPrice')

    # Mengurutkan invoice price dari terbesar ke terkecil
    Invoice_price = data['Quantity'] * data['UnitPrice']
    sorted_invoice_prices = data.assign(Invoice_price=data['UnitPrice']).sort_values(by='Invoice_price', ascending=False)

    # Mengembalikan hasil
    return {
        "Jumlah unik customer": unique_customers,
        "Jumlah unik invoice number": unique_invoices,
        "Jumlah quantity": total_quantity,
        "Jumlah revenue": total_revenue,
        "Jumlah unik deskripsi": unique_descriptions,
        "Price terkecil ke terbesar": sorted_prices,
        "Invoice price terbesar ke terkecil": sorted_invoice_prices
    }

# Memanggil fungsi dengan dataset Anda sebagai parameter
hasil_summary = summarize_data(df)

# Menampilkan jumlah unik customer
print("Jumlah unik customer:", hasil_summary["Jumlah unik customer"])

# Menampilkan jumlah unik invoice number
print("Jumlah unik invoice number:", hasil_summary["Jumlah unik invoice number"])

# Menampilkan jumlah quantity
print("Jumlah quantity:", hasil_summary["Jumlah quantity"])

# Menampilkan jumlah revenue
print("Jumlah revenue:", hasil_summary["Jumlah revenue"])

# Menampilkan jumlah unik deskripsi
print("Jumlah unik deskripsi:", hasil_summary["Jumlah unik deskripsi"])

# Menampilkan 5 data terkecil dari harga unit
print("5 data terkecil dari harga unit:")
print(hasil_summary["Price terkecil ke terbesar"].head())

# Menampilkan 5 data terbesar dari invoice price
print("5 data terbesar dari invoice price:")
print(hasil_summary["Invoice price terbesar ke terkecil"].head())
```

28. Buatlah recency, frequency, dan monetary pada dataset online retail!

*   recency menunjukkan waktu interaksi terakhir pelanggan dengan produk.
*   Frequency mengacu pada berapa kali pelanggan berinteraksi dengan produk atau melakukan transaksi dalam periode waktu tertentu
* Monetary value menunjukkan jumlah total yang dihabiskan oleh pelanggan untuk membeli produk Anda dalam periode waktu tertentu.

```python
# Hitung recency
recency = dataset_cleaned.groupby('CustomerID')['InvoiceDate'].max()
last_invoice_date = dataset_cleaned['InvoiceDate'].max()
recency = (last_invoice_date - recency).dt.days

# Hitung frequency
frequency = dataset_cleaned.groupby('CustomerID')['InvoiceNo'].nunique()

# Hitung monetary value
monetary = dataset_cleaned.groupby('CustomerID')['Revenue'].sum()
```

29. ubahlah nama kolom pada data_rfm menjadi recency,frequency, dan monetory

```python
# Gabungkan recency, frequency, dan monetary value menjadi satu dataframe baru
data_rfm = pd.DataFrame({'Recency': recency, 'Frequency': frequency, 'Monetary': monetary})

# Tampilkan 5 baris teratas dari dataset baru
data_rfm.head()
```

33. Bagaimana cara mengklasifikasi Quantity <= 6 True dan lainnya False?

36. Bagaimana menghitung country yang memiliki top 10 quantity?

```python
# Mengelompokkan data berdasarkan negara dan menjumlahkan quantity untuk setiap negara
country_quantity = dataset_cleaned.groupby('Country')['Quantity'].sum()

# Mengambil 10 negara dengan jumlah quantity terbesar
top_10_countries_quantity = country_quantity.nlargest(10)

# Menampilkan top 10 negara dengan jumlah quantity terbesar
print("Top 10 negara dengan jumlah quantity terbesar:")
print(top_10_countries_quantity)
```

37. Bagaimana menghitung country yang memiliki top 10 revenue?

```python
# Mengelompokkan data berdasarkan negara dan menjumlahkan revenue untuk setiap negara
country_revenue = dataset_cleaned.groupby('Country')['Revenue'].sum()

# Mengambil 10 negara dengan jumlah revenue terbesar
top_10_countries_revenue = country_revenue.nlargest(10)

# Menampilkan top 10 negara dengan jumlah revenue terbesar
print("Top 10 negara dengan jumlah revenue terbesar:")
print(top_10_countries_revenue)
```

38. Ambilah dataset yang memiliki kriteria top 10 pada quantity dan revenue!

```python
# Mengambil negara yang merupakan gabungan dari top 10 quantity dan top 10 revenue
top_countries = top_10_countries_quantity.index.union(top_10_countries_revenue.index)

# Filter dataset asli berdasarkan negara-negara yang terpilih
dataset_top_countries = dataset_cleaned[dataset_cleaned['Country'].isin(top_countries)]

# Menampilkan 10 baris teratas dari dataset yang telah difilter
print("Dataset yang memiliki kriteria top 10 pada quantity dan revenue:")
dataset_top_countries.head(10)
```

39. Bagaimana pertumbuhan quantity dan revenue dari tahun ke tahun pada dataset nomor 38?

```python
# Ekstrak tahun dari kolom 'InvoiceDate'
dataset_top_countries['Year'] = dataset_top_countries['InvoiceDate'].dt.year

# Hitung total quantity dan revenue untuk setiap tahun
total_quantity_revenue_yearly = dataset_top_countries.groupby('Year').agg({'Quantity': 'sum', 'Revenue': 'sum'})
total_quantity_revenue_yearly
```

```python
# Hitung pertumbuhan quantity dan revenue dari tahun ke tahun
pertumbuhan_quantity = total_quantity_revenue_yearly['Quantity'].diff()
pertumbuhan_revenue = total_quantity_revenue_yearly['Revenue'].diff()

# Tampilkan hasil
print("Pertumbuhan quantity dari tahun ke tahun:")
print(pertumbuhan_quantity)
print("\nPertumbuhan revenue dari tahun ke tahun:")
print(pertumbuhan_revenue)
```
